Seaborn is a statistical data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn is particularly useful for visualizing complex datasets with multiple variables. It comes with several built-in themes and color palettes to make it easy to create aesthetically pleasing visualizations

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# importing data from google drive
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Pythonclass/Pokemon.csv', index_col=0, encoding = 'unicode-escape')
```

```python
df.head()
```

Scatter plot

```python
sns.lmplot(x="Attack",y="Defense",data=df)
plt.show()
```

```python
sns.lmplot(x="Attack",y="Defense",data=df, fit_reg=False, hue='Stage')
plt.show()
```

Box plot

```python
df_copy = df.drop(['Total','Stage','Legendary'],axis=1)
```

```python
sns.boxplot(data=df_copy)
```

Violin plot


```python
sns.violinplot(data=df_copy)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="Type 1", y="Attack", data=df_copy)
plt.show()
```

HeatMap

```python
corr = df_copy.corr()

sns.heatmap(corr)
```

Histogram

```python
sns.distplot(df.Attack, color='blue')
```

Count plot

```python
sns.countplot(x="Type 1", data=df)
plt.xticks(rotation=45)
```

Density plot

```python
sns.jointplot(x='Attack', y='Defense', data=df, kind='kde', color='lightblue')
```
