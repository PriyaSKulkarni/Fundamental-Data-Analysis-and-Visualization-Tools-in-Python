Plotly is an open-source data visualization library that allows users to create interactive plots and dashboards in Python, R, and Julia. It supports a wide range of chart types, including scatter plots, line charts, bar charts, histograms, 3D plots, geographic maps, and more. One of the key features of Plotly is its ability to generate interactive visualizations, enabling users to explore and interact with the data directly within the generated plots.

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# importing data from google drive
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Pythonclass/Pokemon.csv', index_col=0, encoding = 'unicode-escape')
```

```python
df.head()
```

Scatter plot

```python
fig = px.scatter(df, x='Attack',y="Defense")
fig.show()
```

```python
fig = px.scatter(df, x='Attack',y="Defense", color="Stage",hover_data=['Name'])
fig.show()
```

```python
fig = px.scatter(df, x='Attack',y="Defense", color="Type 1",size='Total' ,hover_data=['Name'])
fig.show()
```

Line plots:

```python
x=list(range(df.shape[0]))

y= df['Total'].sort_values()

fig =px.line(df,x,y,labels={'x':'Pokemon rank','y':'Total Score'},hover_data=['Name'])
fig.show()
```

Box Plot

```python
df_copy = df.drop(['Name','Type 1','Type 2','Total','Stage','Legendary'],axis =1)
fig =px.box(df_copy)
fig.show()
```

Violin plot

```python
fig = px.violin(df_copy)
fig.show()
```

```python
fig = px.violin(df, x ="Type 1", y="Attack", color="Type 1")
fig.show()
```

HeatMap:

```python
corr = df_copy.corr()
labels=list(df_copy)
fig = px.imshow(corr , x=labels, y=labels)
fig.show()
```

Histogram:

```python
fig= px.histogram(df, x="Attack")
fig.show()
```

Density Heatmap

```python
fig = px.density_heatmap(df,x="Attack",y="Defense")
fig.show()
```
